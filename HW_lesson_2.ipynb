## Урок 2. Анализ данных и проверка статистических гипотез.

### Домашнее задание

#### 1. В чём различие между зависимыми и независимыми выборками?

### Решение:

*В случае с* **зависимыми** *выборками выбор объекта из одной выборки влияет на выбор из другой (на его вероятность).*

*Например, одна и та же выборка до и после какого-то воздействия.*

*В случае с* **независимыми** *выборками вероятность выбора объекта из одной выборки не влияет на выбор объекта из другой.*

*Например, мужчины и женщины в каком-то списке.*

#### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

### Решение:

*Параметрические статистические критерии применяются для нормального и близких к нормальному распределений.*

*Непараметрические аналоги - в случаях распределений, отличных от нормального.*
